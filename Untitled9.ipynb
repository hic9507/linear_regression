csv 파일 생성

```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

!pip install yfinance
import yfinance as yf
from datetime import datetime, timedelta
import random

def createCSVdata(startDate, endDate, dataLength, dataNumber, fileName):
    start = datetime.strptime(startDate, '%Y-%m-%d')
    end = datetime.strptime(endDate, '%Y-%m-%d')

    # 전체 기간을 일자 리스트로 변환
    date_generated = [start + timedelta(days=x) for x in range(0, (end-start).days)]

    # 데이터 프레임 생성
    data = {'Date': [], 'Value': []}

    # dataNumber만큼 데이터 생성
    for _ in range(dataNumber):
        # 시작일자부터 dataLength만큼의 연속 일자 선택
        random_start = random.choice(date_generated[:-dataLength])
        date_range = [random_start + timedelta(days=x) for x in range(dataLength)]

        # 해당 일자에 대한 무작위 값 생성
        values = np.random.random(dataLength)

        # 데이터 프레임에 추가
        for date, value in zip(date_range, values):
            data['Date'].append(date)
            data['Value'].append(value)

    # 데이터 프레임 생성 및 파일로 저장
    df = pd.DataFrame(data)
    df.to_csv(fileName, index=False)
    print(f'File saved as {fileName}')

# 함수 사용 예시
createCSVdata('2022-11-01', '2022-11-15', 5, 10, 'sample_data.csv')
```
